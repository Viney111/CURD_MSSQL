# [dbo].[sp_insert_record_tblcustomerinfo]

```sql
/****** Object:  StoredProcedure [dbo].[sp_insert_record_tblcustomerinfo]    Script Date: 04-May-22 12:52:22 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_insert_record_tblcustomerinfo] (
@Name VARCHAR(20),
@Address VARCHAR (30),
@Phone BIGINT,
@Email  VARCHAR(30),
@Zip INT
)
AS
BEGIN
INSERT INTO customer_info(Name,Address,Phone,Email,Zip)
VALUES(@Name,@Address,@Phone,@Email,@Zip)
END
GO
```

# [dbo].[sp_update_record_tblcustomerinfo]

```sql
/****** Object:  StoredProcedure [dbo].[sp_update_record_tblcustomerinfo]    Script Date: 04-May-22 12:52:22 PM ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[sp_update_record_tblcustomerinfo] (
@ID INT,
@Name VARCHAR(20),
@Address VARCHAR (30),
@Phone BIGINT,
@Email  VARCHAR(30),
@Zip INT
)
AS
BEGIN
UPDATE customer_info
SET
Name = @Name,
Address = @Address,
Phone = @Phone,
Email = @Email,
Zip = @Zip
WHERE ID = @ID
END
GO
```
